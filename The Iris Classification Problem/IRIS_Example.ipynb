PREPARE INPUT DATA FOR DEEP LEARNİNG

perform the following steps for preparing data

 1. Load data into a pandas dataframe
 2. Convert the dataframe to a numpy array
 3. Use one-hot-encoding for the target variable
 4. Split into training and tests datasets

```python
import pandas as pd
import os
import tensorflow as tf
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Load the data and review content
iris_data = pd.read_csv("iris.csv")


#print("\nLoaded Data : n\------------------------")
print(iris_data.head())

```

```python
type(iris_data["Sepal.Length"])
```

```python
#Use a Label encoder to convert String to numeric  values
# for the Target variable

from sklearn import preprocessing    # used to make LabelEncoder
label_encode = preprocessing.LabelEncoder()

iris_data['Species'] = label_encode.fit_transform(iris_data["Species"])
 # yukarıdkai kod iris_data veri setindeki "Species" sütununu "LabelEncoder" kullanarak sayılsal değerlere dönüştürür.
 # örneğin Species sütununda versicolor, virginica ve setosa gibi veriler var 
 # bu verilere setosa için 0, versicolor için 1 ve virginica için 2 değerlerini atayabilir.
```

```python
iris_data.head() #görüleceği üzere setosa değerlerine 0 atandı
```

```python
iris_data.tail(5) # virginica değerlerine ise 2 atandı
```

```python
# Convert input to numpy array
np_iris=iris_data.to_numpy()  # pandas olan "iris_data" setini numpy dizisine çevirir ve "np_iris" içine atar
```

```python
# Seperate input to numpy array
x_data = np_iris[:,0:4] # x_data, np_iris dizisinin tüm satırlarını (:) ve ilk dört sütununu (0:4) seçer.
y_data = np_iris[:,4]   # y_data, np_iris dizisinin tüm satırlarını (:) ve beşinci sütununu (4) seçer.
```

```python
print(x_data)
```

```python
print(y_data)
```

```python
#Create a scaler model that is fit on the input data.
scaler = StandardScaler().fit(x_data)
```

```python
# Scale the numeric feature variables
x_data = scaler.transform(x_data)
```

```python
x_data #scaler edilmiş hali
```

```python
#Convert target variable as a one-hot-encoding array
y_data = tf.keras.utils.to_categorical(y_data,3)

#yukarıda one-hot-encoding yapmamızın sebebi "y_data" da 0 ,1 ve 2 gibi veriler var
# bu veriler LabelEncoder ile setosa versicolor ve virginica verilerinden sayısal değere çevrilmişti
# yani setosa = 0, versicolor = 1, virginica = 2 ye çevrilmişti
# One-hot-Encoding ile verileri vektör ile temsil edeceğiz
# örneğin setosa = [1, 0, 0]
        # versicolor = [0, 1, 0]
        # virginica = [0, 0, 1] gibi
```

```python
y_data
```

```python
x_data[:5, :] # ilk 5 satır ve bütün sütunları
```

```python
y_data[:5, :]
```

```python
# Split training and test data
x_train, x_test, y_train, y_test = train_test_split ( x_data, y_data, test_size=0.10)
```

```python
print(x_train.shape, y_train.shape, x_test.shape, y_test.shape)
```

#  Creating a Model
 Create a model in Keras requires defining the following

   1. Number of hidden layers
   2. Number of nodes in each layer
   3. Activation Functions
   4. Loss function & Accuracy measurements



```python
from tensorflow import keras
```

```python
# Number of classes in target variable
# there are 3 classes ( setosa, virginica, versicolor)
NB_CLASSES = 3
```

```python
# Create a sequencial model in Keras
model = tf.keras.models.Sequential()
```

```python
# Add the first hidden layer
model.add(keras.layers.Dense (128,                            #Number of nodes
                              input_shape = (4,),             #Number of input variables, there are 4 "x_data" column
                              name = "Hidden-Layer-1",        #Logical name
                              activation = "relu"))           #activation function
```

```python
# Add a second hidden layer
model.add(keras.layers.Dense(128,
                             name = "Hidden-Layer-2",
                             activation="relu"))
```

```python
# Add an Output layer with softmax activation ( softmax activation çok sınıflı sınıflandırmada kulanılır)
model.add(keras.layers.Dense(NB_CLASSES,
                             name = "Output-Layer",
                             activation = "softmax"))
```

```python
# Compile the model with loss & metrics
model.compile(loss="categorical_crossentropy", # çok sınıflı sınıflandırma problemleri için yaygın olarak kullanılır.
              metrics = ["accuracy"])
```

```python
# print the model meta-data
model.summary()
```

# Training and Evaluating the Model
 Training the odel involves defining training models and then perform forward and back propagation

```python
# Make it verbose so we can see the progress
VERBOSE=1
```

yukarıdaki kod bloğunda VERBOSE=1 ifadesi, programın veya modelin çalışırken daha fazla ayrıntılı bilgi vermesini sağlar. Genellikle, bir model eğitilirken veya bir algoritma çalıştırılırken, bu ayar, kullanıcıya eğitim süreci, kayıp değerleri, doğruluk oranları gibi ilerleme ile ilgili detayları göstermek için kullanılır. VERBOSE değişkeni 1 olarak ayarlandığında, çıktılar daha ayrıntılı ve sık olacak şekilde yapılandırılır, bu da kullanıcıya işlem sırasında ne olduğunu daha iyi anlamasına yardımcı olur.

```python
# Setup Hyper Parameters for training

# Set Batch Size
BATCH_SIZE = 16   #modelin eğitilmesi aşamasında aynı anda kaç adet verinin işleneceği anlamına gelir
# Set number of epochs
EPOCHS = 10
# Set validation split. 20% of the training data will be used for vaidation
# after each epoch
VALIDATION_SPLIT = 0.2   # önerilen %10 dur ancak bizim verimiz az bu yüzden %20 kullanıyorum.
```

```python

print("\nEğitim İlerlemesi:\n------------------------------------")

# Fit the model. This will perform the entire training cycle, including
# forward propagation, loss computation, backward propagation and gradient descent.
# Execute for the specified batch size and epoch
# Perform validation after each epoch
history = model.fit(x_train,
                    y_train,
                    batch_size = BATCH_SIZE,
                    epochs= EPOCHS,
                    verbose = VERBOSE,
                    validation_split = VALIDATION_SPLIT)
```

```python
print("\nEğitim Sırasında Doğruluk :\n------------------------------------")

import matplotlib.pyplot as plt

# plot accuracy of the model after each epoch.
pd.DataFrame(history.history)["accuracy"].plot(figsize=(8,5))
plt.title("Accuracy improvements with Epoch")
plt.show()

# Evaluate the model against the test dataset and print results
print("\nTest Veri Kümesine Karşı Değerlendirme :\n------------------------------------")
model.evaluate(x_test,y_test)
```
